```python
import pandas as pd
```

```python
# Membuat dataframe dari data berupa file excel
df = pd.read_excel('midfielderstop5leagues.xlsx')
```

```python
df.tail()
```

## Menambahkan statistik permainan Casemiro pada musim terbaiknya di Manchester United (musim 2022/2023) ke dalam dataframe.

```python
stat_casemiro_2022_2023 = {
    'Player': 'Casemiro',
    'Passes Attempted': 1447,
    'Pass Completion %': 77.5,
    'Progressive Passes': 152,
    'Tackles': 88,
    'Blocks': 49,
    'Clearances': 64,
    'Interceptions': 34,
    'Aerial Duels Won': 56
}

# Menggunakan loc untuk menambahkan baris baru
df.loc[len(df)] = stat_casemiro_2022_2023
```

```python
df.tail()
```

## Data preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Pilih kolom yang relevan untuk clustering
features = ['Passes Attempted', 'Progressive Passes', 'Tackles', 'Blocks', 'Clearances', 'Interceptions', 'Aerial Duels Won', 'Pass Completion %']
```

```python
# Normalisasi data
scaler = StandardScaler()
df_normalized = scaler.fit_transform(df[features])
```

```python
# Tentukan jumlah kluster (misalnya, 10 kluster)
kmeans = KMeans(n_clusters=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_normalized)
```

```python
# Menemukan kluster tempat Casemiro berada
casemiro_cluster = df[df['Player'] == 'Casemiro']['Cluster'].values[0]
```

```python
# Filter pemain yang berada di kluster yang sama dengan Casemiro
similar_players = df[df['Cluster'] == casemiro_cluster]
```

```python
# Menampilkan pemain yang mirip dengan Casemiro

# Mengatur opsi tampilan Pandas
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.expand_frame_repr', False)  # Mencegah pemisahan tabel

# Menampilkan pemain yang mirip dengan Casemiro dalam bentuk tabel
display(similar_players[['Player', 'Cluster'] + features])
```
